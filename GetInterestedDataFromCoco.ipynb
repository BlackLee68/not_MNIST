```python
def judge_function(index_list, line):
    temp = line.strip().split(" ")
    if int(temp[0]) in index_list:
        return True
    else:
        return False
```

```python
# here we delete information in labeled txt files
root_path = "/usr1/darknet/darknet-master/data/coco/labels/"
k = 0
for father_name in ["val2014","train2014"]:
    path = root_path + father_name + "/"
    if father_name == "val2014":
        for val_name in labeld_val_txt_files:
            i = 0
            k += 1
            #print(val_name)
            with open(path + val_name, 'r') as val_file:
                lines = val_file.readlines()
            with open(path + val_name, 'w') as val_file2:
                #i = 0
                for l in lines:
                    if judge_function(target_index, l):
                        val_file2.write(l)
                        i += 1
            if i == 0:
                # delete the txt file
                os.remove(path + val_name)
            if k % 1000 == 0:
                print(k)
    else:
        for train_name in labeld_train_txt_files:
            j = 0
            k += 1
            with open(path + train_name, 'r') as train_file:
                lines_train = train_file.readlines()
            with open(path + train_name, 'w') as train_file2:
                #i = 0
                for l_tra in lines_train:
                    if judge_function(target_index, l_tra):
                        train_file2.write(l_tra)
                        j += 1
            if j == 0:
                # delete the txt file
                os.remove(path + train_name)
            if k % 1000 == 0:
                print(k)
```

```python
def map_function(string):
    a = [0,1,2,3,5,7,16,17,18,19,20,21,32,39,57,58]
    return str(a.index(int(string)))
```

```python
def make_newline(new_str_list):
    string = ""
    for item in new_str_list:
        string += item + " "
    return string
```

```python
# make new .names file, make a map from original to new class index the labels/* should be changed
root_path = "/usr1/darknet/darknet-master/data/coco/labels/"
k = 0
for father_name in ["val2014","train2014"]:
    path = root_path + father_name + "/"
    if father_name == "val2014":
        for val_name in labeld_val_txt_files:
            k += 1
            #print(val_name)
            with open(path + val_name, 'r') as val_file:
                lines = val_file.readlines()
            with open(path + val_name, 'w') as val_file2:
                i = 0
                for l in lines:
                    old_index_list = l.strip().split(" ")
                    old_index = old_index_list[0]
                    new_index = map_function(old_index)
                    old_index_list[0] = new_index
                    new_line = make_newline(old_index_list)
                    if i != len(lines) - 1: 
                        val_file2.write(new_line+"\n")
                    else:
                        val_file2.write(new_line)
                    i += 1
            if k % 1000 == 0:
                print(k)
    else:
        for train_name in labeld_train_txt_files:
            k += 1
            with open(path + train_name, 'r') as train_file:
                lines_train = train_file.readlines()
            with open(path + train_name, 'w') as train_file2:
                j = 0
                for l_tra in lines_train:
                    old_index_list = l_tra.strip().split(" ")
                    old_index = old_index_list[0]
                    new_index = map_function(old_index)
                    old_index_list[0] = new_index
                    new_line = make_newline(old_index_list)
                    if j != len(lines) - 1: 
                        train_file2.write(new_line+"\n")
                    else:
                        train_file2.write(new_line)
                    j += 1
            if k % 1000 == 0:
                print(k)
```
